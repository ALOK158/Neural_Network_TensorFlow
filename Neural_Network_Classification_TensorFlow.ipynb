 To visulize our modeles predictions, lets create a function
plots , this function will:

 Take in a trained model, features (X) and labels (y)

 Create a meshgrid of the different X values

 Make predictions across the meshgrid

 PLot the predictions as  well as line bewtween different zones


```python
import numpy as np
def plot_decision_boundary(model,x,y):
  """
  plot the decicion boundary created by model predicitons on X
  """
  x_min,x_max=x[:,0].min()-0.1,x[:,0].max()+0.1
  y_min,y_max=x[:,1].min()-0.1,x[:,1].max()+0.1

  ##meshgrid
  xx, yy=np.meshgrid(np.linspace(x_min,x_max,100),
                    np.linspace(y_min, y_max,100)
  )

  #create x values
  x_in=np.c_[xx.ravel(),yy.ravel()]

  #create predictions
  y_pred=model.predict(x_in)

  if(len(y_pred))>1:
    print("multi-class")
    # we have to reshape the predictions
    y_pred=np.argmax(y_pred,axis=1).reshape(xx.shape)
  else:
    print("binary")
    y_pred=np.round(y_pred).reshape(xx.shape)

  #plot the graph
  plt.contourf(xx,yy, y_pred,cmap=plt.cm.RdYlBu, alpha=0.7)
  plt.scatter(x[:,0],x[:,1],c=y,s=40,cmap=plt.cm.RdYlBu)
  plt.xlim(xx.min(), xx.max())
  plt.show()
```

```python
#check out the predciotn our model in making
plot_decision_boundary(model,x,y)
```

###Finding the best learning rate

To find the ideal learning rate(the learning rate where losss reduces most during the training ) we are going to use following steps:

* A learning rate **callback**- extra piece of funcionality , can add to your model while traning
* A modified loss curves plot

```python


# set seed
tf.random.set_seed(42)

#Create the model

model_x= tf.keras.Sequential([
        tf.keras.layers.Dense(32,activation="relu"),
        tf.keras.layers.Dense(8,activation="relu"),
        tf.keras.layers.Dense(2,activation="relu"),
        tf.keras.layers.Dense(1,activation="sigmoid")
])

# compile the model
model_x.compile(loss="binary_crossentropy",
                optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
                metrics=["accuracy"])
# create the learning rate call backs
lr_scheduler=tf.keras.callbacks.LearningRateScheduler(lambda epoch: 1e-4*10**(epoch//20) )

#Fit the model

history_x=model_x.fit(x_train,y_train,epochs=50,callbacks=[lr_scheduler])
```

# Building multi class classification model
We can use similar model as our binary classifier with some tuning

      * Input shape-(28,28)
      * Output shape 10
      * lossfucntion
      * Output Activation fucntion

```python
# set random seed
tf.random.set_seed(42)

# Create the model
model_x=tf.keras.Sequential([
        tf.keras.layers.Flatten(input_shape=(28,28)),
        tf.keras.layers.Dense(4, activation="relu"),
        tf.keras.layers.Dense(10, activation="softmax")
])
# compile the model
model_x.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(),
                optimizer=tf.keras.optimizers.Adam(),
                metrics=["accuracy"])

#Fit the model
h_x=model_x.fit(train_data,train_labels,epochs=10, validation_data=(test_data, test_labels))
```

key takeaways

        *CrossEntropy- works on when desired output is onehot encoded

        *If labels are in integers form use SparseCrossEntropy

How about we create a fun little fucntion for :

*   Plot a random image
*   Make predixtion on said image

* lavel the plot wiht the truth label & the predixted label

```python
def plot_random_image(model, images, true_labels, classes):
  """
  Picks a random image, plots it and labels it with a prediction and trutg label
  """
  # set up random iteger
  i=random.randint(0, len(images))

  # Create predictions and targets
  target_image=images[i]
  pred_probs=model.predict(target_image.reshape(1,28,28))
  pred_label=class_names[pred_probs.argmax()]
  true_label=class_names[true_labels[i]]

  #plot the image
  plt.imshow(target_image, cmap=plt.cm.binary)

  # change the color of titles dependind on if predcition is right or wrong

  if pred_label==true_label:
    color="green"
  else:
    color="red"

  ## Add xlael
  # Remove spaces within the curly braces and use correct format specifiers
  plt.xlabel( "pred: {} {:2.0f}% (True: {}) " .format(pred_label,
                100 *tf. reduce_max(pred_probs) ,
                true_label),
                color=color) # set the color to green or red based on if prediction i:


```
