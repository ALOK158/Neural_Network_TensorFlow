 To visulize our modeles predictions, lets create a function
plots , this function will:

 Take in a trained model, features (X) and labels (y)

 Create a meshgrid of the different X values

 Make predictions across the meshgrid

 PLot the predictions as  well as line bewtween different zones


```python
import numpy as np
def plot_decision_boundary(model,x,y):
  """
  plot the decicion boundary created by model predicitons on X
  """
  x_min,x_max=x[:,0].min()-0.1,x[:,0].max()+0.1
  y_min,y_max=x[:,1].min()-0.1,x[:,1].max()+0.1

  ##meshgrid
  xx, yy=np.meshgrid(np.linspace(x_min,x_max,100),
                    np.linspace(y_min, y_max,100)
  )

  #create x values
  x_in=np.c_[xx.ravel(),yy.ravel()]

  #create predictions
  y_pred=model.predict(x_in)

  if(len(y_pred))>1:
    print("multi-class")
    # we have to reshape the predictions
    y_pred=np.argmax(y_pred,axis=1).reshape(xx.shape)
  else:
    print("binary")
    y_pred=np.round(y_pred).reshape(xx.shape)

  #plot the graph
  plt.contourf(xx,yy, y_pred,cmap=plt.cm.RdYlBu, alpha=0.7)
  plt.scatter(x[:,0],x[:,1],c=y,s=40,cmap=plt.cm.RdYlBu)
  plt.xlim(xx.min(), xx.max())
  plt.show()
```

```python
#check out the predciotn our model in making
plot_decision_boundary(model,x,y)
```

**Note**: The combiantion of **linear** and **non linear** fucntions is one of key fundamentals for neural newtwork

```python
#Create a toy tensor (similar to data we pass into our models)

A=tf.cast(tf.range(-10,10), tf.float32)
A
```

```python
# lets start replicating sigmoid=1/(1+exp(-x))
def sigmoid(x):
  return 1/(1+tf.exp(-x))

#use sigmoid on A
sigmoid(A)


```

```python
plt.plot(sigmoid(A))
```

```python
#Having relu fucntion
def relu(x):
  return tf.maximum(x,0)
plt.plot(relu(A))
```
